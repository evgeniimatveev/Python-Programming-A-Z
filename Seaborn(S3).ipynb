# EXAMPLE #1

```python
# üìö Import essential libraries  
import pandas as pd  # For handling tabular data üêº  
import numpy as np  # For numerical operations üî¢  
import matplotlib.pyplot as plt  # For visualization üé®  
import seaborn as sns  # For statistical data visualization üìä  

# üî¨ Import the dataset from scikit-learn  
from sklearn.datasets import load_breast_cancer  
```

```python
# üî¨ Load the Breast Cancer dataset  
cancer = load_breast_cancer()  

# üèóÔ∏è Convert the dataset into a Pandas DataFrame  
df_cancer = pd.DataFrame(
    np.c_[cancer['data'], cancer['target']],  # Combine feature data with target  
    columns=np.append(cancer['feature_names'], ['target'])  # Assign column names  
)  

# üèÜ Display the first few rows for an overview  
df_cancer.head()  
```

```python
# üé≠ Seaborn Pairplot ‚Äì Visualizing Feature Relationships  
sns.pairplot(df_cancer,  
             hue="target",  # üé® Color by class (Malignant = Orange, Benign = Blue)  
             vars=['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'],  
             palette="coolwarm",  # üåà Beautiful color mapping  
             markers=["o", "s"],  # üîµ Use 'o' for benign, üü† 's' for malignant  
             diag_kind="hist")  # üìä Histograms on diagonal for better readability  

# üèÜ Display the plot  
plt.show()  
```

```python
# üìä Corrected Countplot: Assign x explicitly and disable legend  
sns.countplot(x=df_cancer['target'], hue=df_cancer['target'], palette="coolwarm", legend=False)  

# üéØ Display the plot  
plt.show()
```

```python
# üé≠ Enhanced Scatterplot: Adding size, style & legend  
plt.figure(figsize=(8,6))  # üîç Adjust figure size for better clarity  
sns.scatterplot(x='mean area', y='mean smoothness', hue='target', data=df_cancer, palette="coolwarm",  
                style="target", size="mean perimeter", sizes=(10, 100))  # üéØ Different styles & sizes  
plt.title("Mean Area vs Mean Smoothness with Size Representation")  # üèÜ Add a title  
plt.xlabel("Mean Area (Size of Tumor)")  # üè∑ Label X-axis  
plt.ylabel("Mean Smoothness")  # üè∑ Label Y-axis  
plt.legend(title="Cancer Type", labels=["Malignant (0)", "Benign (1)"])  # üé® Improved legend  
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# üìè **Set figure size for better readability**
plt.figure(figsize=(17, 10)) 

# üé® **Create an enhanced heatmap with better visualization**
sns.heatmap(
    df_cancer.corr(),  # üìä Compute correlation matrix
    annot=True,  # üìù Display correlation values inside cells
    fmt=".2f",  # üî¢ Format numbers to 2 decimal places
    cmap="coolwarm",  # üé® Use color theme for better contrast
    linewidths=1,  # ‚ûñ Add grid lines between cells
    square=False,  # üî≤ Ensure rectangular aspect ratio
    cbar_kws={"shrink": 0.8, "label": "Correlation Strength"}  # üéõ Customize color bar
)

# üè∑ **Add title and labels for clarity**
plt.title(" Titanic Feature Correlation Heatmap", fontsize=16, fontweight='bold')  # ‚úÖ –£–≤–µ–ª–∏—á–∏–ª–∏ —Ä–∞–∑–º–µ—Ä –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.xticks(rotation=45, ha='right', fontsize=10)  # üîÑ –£–ª—É—á—à–∏–ª–∏ —á–∏—Ç–∞–µ–º–æ—Å—Ç—å –ø–æ–¥–ø–∏—Å–µ–π
plt.yticks(fontsize=10)  # üìè –°–¥–µ–ª–∞–ª–∏ –ø–æ–¥–ø–∏—Å–∏ –Ω–∞ –æ—Å–∏ Y —á–µ—Ç—á–µ

# üñº **Display the heatmap**
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# üéØ Load Titanic dataset
titanic_data = sns.load_dataset("titanic")

# üè∑Ô∏è Display first 5 rows to understand the data
titanic_data.head()
```

```python
# üö¢ Count of passengers in each class
plt.figure(figsize=(8, 5))  # üìè Adjust figure size
sns.countplot(x="class", hue="class", data=titanic_data, palette="viridis", legend=False)  # üé® Use color palette

plt.xlabel("Passenger Class", fontsize=12)  # üè∑Ô∏è Label x-axis
plt.ylabel("Count", fontsize=12)  # üè∑Ô∏è Label y-axis
plt.title("Titanic Passenger Class Distribution", fontsize=14)  # üìä Standard title without emoji
plt.show()  # üéâ Display the plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# üéØ Survival distribution across passenger classes
plt.figure(figsize=(8, 5))  # üìè Adjust figure size

# üõü Using 'hue' to fix the warning and improve clarity
sns.countplot(x="survived", hue="survived", data=titanic_data, palette="coolwarm", legend=False)

# üè∑Ô∏è Label axes for clarity
plt.xlabel("Survival (0 = No, 1 = Yes)", fontsize=12)  # Label x-axis
plt.ylabel("Count", fontsize=12)  # Label y-axis
plt.title("Titanic Survival Count", fontsize=14)  # üèõÔ∏è Add title without emoji to avoid font issues

# üé® Improve visual aesthetics
plt.xticks([0, 1], ["Did Not Survive", "Survived"])  # Custom x-axis labels
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Light grid for readability

plt.show()  # üéâ Display the plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# üì• Load Titanic dataset
titanic_data = sns.load_dataset("titanic")

# üé® Set Seaborn style for better visualization
sns.set_style("whitegrid")
```

```python
titanic_data
```

```python
# üì¶ **Boxplot**: Shows median, quartiles, and outliers
plt.figure(figsize=(8, 5))  # üîπ Adjust figure size
sns.boxplot(x='class', y='age', data=titanic_data, hue='class', palette="Set2")  # ‚úÖ FIXED: Added `hue`

# üè∑Ô∏è Labels and title
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.title(" Titanic Age Distribution (Boxplot)", fontsize=14)

plt.show()  # üîπ Display the plot
```

```python
# üêù **Swarmplot**: Individual data points (less overlap)
plt.figure(figsize=(8, 5))  # üîπ Adjust figure size
sns.swarmplot(x='class', y='age', data=titanic_data, hue='class', palette="Set1", size=2.5, alpha=0.7, dodge=False)  
# ‚úÖ FIXED: `size=2.5` reduces marker size, `dodge=False` prevents overlap issues

# üè∑Ô∏è Labels and title
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.title(" Titanic Age Distribution (Swarmplot)", fontsize=14)

plt.show()  # üîπ Display the plot
```

```python
import seaborn as sns  # üìä Import seaborn for data visualization
import matplotlib.pyplot as plt  # üìà Import matplotlib for plotting
import pandas as pd  # üóÇ Import pandas for data manipulation

# üìå Load Titanic dataset
titanic_data = sns.load_dataset("titanic")

# üõë Drop non-numeric columns for correlation matrix to work properly
numeric_data = titanic_data.select_dtypes(include=['number'])

```

```python
titanic_data
```

```python
# üì¶ Boxplot: Shows distribution of age across passenger classes
sns.boxplot(x="class", y="age", data=titanic_data)
plt.xlabel("Passenger Class")  # üî† Label x-axis
plt.ylabel("Age")  # üî† Label y-axis
plt.title(" Titanic Age Distribution (Boxplot)")  # üè∑Ô∏è Add title
plt.show()  # üëÄ Display the plot
```

```python
# üìä Jointplot: Shows relation between fare and age with marginal histograms
sns.jointplot(x="fare", y="age", data=titanic_data)
plt.show()  # üëÄ Display the plot
```

```python
# üî• Heatmap: Displays correlation between numerical variables
plt.figure(figsize=(10, 8))  # üìè Set figure size for better readability
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")  # üé® Add color mapping
plt.title(" Titanic Feature Correlation Heatmap")  # üè∑Ô∏è Add title
plt.show()  # üëÄ Display the plot
```
