```python
# 📊 MATPLOTLIB BASICS: CREATING BAR CHARTS WITH LOG SCALE 📊
# In this script, we'll explore Matplotlib, a powerful visualization library for Python.
# We'll generate a **bar chart** with logarithmic scaling using `matplotlib.pyplot` and `numpy`.

# 🔹 Importing Necessary Libraries
import matplotlib.pyplot as plt  # Matplotlib for visualization 📊
import numpy as np  # NumPy for numerical operations 🔢

# 🔹 Defining Data
# Our data consists of multiple tuples, each containing (x, y) values
data = [(0, 1000), (1, 30), (2, 100, 30), (3, 500, 800), (4, 50, 1)]

# 🔹 Extracting the number of groups in our dataset
dim = len(data[0])  # Number of elements in each tuple
w = 0.75  # Width of each group bar
dim_w = w / dim  # Adjusting width for multiple bars per x-label

# 🔹 Creating the Plot
fig, ax = plt.subplots()  # Initialize the figure and axis
x = np.arange(len(data))  # Generating x-axis positions

# 🔹 Looping Through Data to Create Bar Groups
for i in range(dim):  # Loop through each dimension in the tuple
    y = [d[i] for d in data]  # Extract the i-th element from each tuple
    ax.bar(x + i * dim_w, y, dim_w, bottom=0.001)  # Plot each bar group

# 🔹 Customizing the Axes
ax.set_xticks(x + dim_w / 2)  # Adjust x-tick positions for better visibility
ax.set_xticklabels(map(str, x))  # Labeling x-axis with numeric values
ax.set_yscale('log')  # Setting **logarithmic scale** for y-axis 📈

# 🔹 Adding Labels and Displaying the Plot
ax.set_xlabel('X')  # Label for the x-axis
ax.set_ylabel('Y')  # Label for the y-axis
plt.show()  # Display the chart 🎨

# 📝 SUMMARY:
# ✅ Used `plt.bar()` to create a **grouped bar chart**.
# ✅ Implemented a **logarithmic scale** for better visualization of large differences.
# ✅ Adjusted **bar width** dynamically to accommodate multiple bars per x-label.
# ✅ Customized **x-axis labels** for clarity.
# ✅ Used `bottom=0.001` to avoid zero-logarithm errors.

# 🔥 Practical Uses:
# 📊 Comparing values across categories with **huge variations** (e.g., financial data, scientific measurements).
# 🚀 Visualizing **growth trends** where differences span multiple orders of magnitude.
```

# BASIC PLOT

```python
# 📈 BASIC PLOTTING WITH MATPLOTLIB 📉
# Matplotlib is a powerful library for data visualization.
# In this section, we'll cover **basic plots**, including:
# - Line plots 📈
# - Scatter plots 🔵
# - Multiple plots on the same figure 📊

# 🔹 Importing Required Libraries
import numpy as np  # NumPy for numerical computations 🔢
import matplotlib.pyplot as plt  # Matplotlib for plotting 🎨

# 🔹 Generating Data for a Basic Line Plot
# We'll create a sine wave 🌊 using NumPy's `arange` and `sin` functions.
t = np.arange(0., 10., 0.2)  # Time interval from 0 to 10, step of 0.2
y = np.sin(t)  # Compute sine values

# 🔹 Creating the Line Plot 📈
plt.plot(t, y, label="Sine Wave", color='b', linestyle='-', marker='o')  
plt.xlabel("Time ⏳")  # Label for X-axis
plt.ylabel("Amplitude 📏")  # Label for Y-axis
plt.title("Basic Line Plot: Sine Wave 🌊")  # Title of the plot
plt.legend()  # Show the legend 📜
plt.grid(True)  # Add grid for better readability 🏁
plt.show()  # Display the plot 🖼️

# 🎯 SUMMARY:
# ✅ Used `plt.plot()` to create a simple **sine wave** plot.
# ✅ Added **labels, title, and grid** for clarity.
# ✅ Included a **legend** to explain the plotted line.

# 🔥 Practical Uses:
# 📊 Line plots are great for **time series analysis**, **stock prices**, and **sensor data**.

# ------------------------------------------------------------

# 🔹 SCATTER PLOT: RANDOM DATA POINTS 🔵
# Scatter plots are useful for **visualizing relationships** between two numerical variables.

# Generate random data points
x = np.random.randn(500)  # 500 random numbers for X-axis
y = np.random.randn(500)  # 500 random numbers for Y-axis

# Creating Scatter Plot
plt.scatter(x, y, color='dodgerblue', alpha=0.6, edgecolors='black')  
plt.xlabel("X Axis ➡️")  
plt.ylabel("Y Axis ⬆️")  
plt.title("Random Scatter Plot 🔵")  
plt.grid(True, linestyle='--', alpha=0.5)  # Dashed grid lines
plt.show()  

# 🎯 SUMMARY:
# ✅ Used `plt.scatter()` to create a **scatter plot**.
# ✅ Adjusted **colors, transparency, and edge color**.
# ✅ Scatter plots are used for **correlation analysis, trend detection, and clustering**.

# ------------------------------------------------------------

# 🔹 MULTIPLE PLOTS ON THE SAME FIGURE 🖼️
# Sometimes, we need to **visualize multiple datasets** in one figure.

# Generating multiple datasets
x = np.linspace(0, 10, 100)  # 100 evenly spaced numbers
y1 = np.sin(x)  # Sine wave
y2 = np.cos(x)  # Cosine wave

# Creating multiple line plots on the same figure
plt.plot(x, y1, label="Sine 🌊", linestyle='-', color='b')
plt.plot(x, y2, label="Cosine 🔄", linestyle='--', color='r')

# Customizing the plot
plt.xlabel("X Values")  
plt.ylabel("Function Values")  
plt.title("Multiple Line Plots 🎭")  
plt.legend()  
plt.grid(True)  
plt.show()  

# 🎯 SUMMARY:
# ✅ Used `plt.plot()` multiple times to **overlay multiple plots**.
# ✅ Differentiated plots using **colors, line styles, and labels**.
# ✅ Useful for **comparing trends, phase shifts, and mathematical functions**.

# 🚀 NEXT STEPS:
# - Try **changing colors** and **line styles** 🎨.
# - Experiment with **different markers** (circle, square, star) 🌟.
# - Explore **subplots** to organize multiple plots efficiently 🖼️.
```

# MULTIPLE PLOTS ON SAME FIGURE/LINE FORMATTING

```python
# 📌 Import necessary libraries  
import numpy as np  # NumPy for numerical operations 🔢  
import matplotlib.pyplot as plt  # Matplotlib for plotting 📈  

# 🔹 Create an array of evenly spaced values from 0 to 5 with step 0.2  
t = np.arange(0.0, 5.0, 0.2)  

# 🎨 Plot multiple lines on the same figure  
# 'r--'  -> Red dashed line  
# 'bs'   -> Blue squares  
# 'g^'   -> Green triangles  
plt.plot(t, t, 'r--',  # Linear plot (y = x) 📉  
         t, t**2, 'bs',  # Quadratic plot (y = x²) 🔵  
         t, t**3, 'g^',  # Cubic plot (y = x³) 🟢  
         linewidth=5.0)  # Set line width to 5 for better visibility  

# 🖼️ Display the plot  
plt.show()
```

# MATPLOTLIB WITH OBJECT ORIENTED PROGRAMMING

- matplotlib creates a Figure instance, which is an object describing the plot window and its properties, and containing lists of all its elements
- matplotlib creates an Axes element within the figure. An axes can be thought of as a plotting window, where data can be arranged by x and y coordinates.
- add_axes(rectangle), where rectangle is a list [x0, y0, width, height] denoting the lower left point of the new axes in figure coodinates (x0,y0) and its width and height. 



```python
# 📌 Step 1: Import necessary libraries
import numpy as np  # NumPy is used for generating data points 🔢
import matplotlib.pyplot as plt  # Matplotlib is for creating visualizations 📈
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# 🔹 Step 2: Generate the time values (0 to 10 with intervals of 0.1)
# This will be the X-axis values for our plots
t = np.arange(0.0, 10.0, 0.1)  

# 🔹 Step 3: Create multiple sine wave data
y1 = np.sin(t)      # Standard sine wave (sin(t)) 🌊
y2 = np.sin(2 * t)  # Sine wave with double the frequency (sin(2t))
y3 = np.sin(3 * t)  # Sine wave with triple the frequency (sin(3t))

# 🎨 Step 4: Create a figure with subplots
# `fig, axs = plt.subplots(rows, cols, figsize=(width, height))`
# Creates a grid of plots, where `axs` is a 2D array of plot axes
fig, axs = plt.subplots(2, 2, figsize=(10, 6))  # 2x2 grid of subplots

# 📈 Step 5: Plot on the first subplot (top-left)
axs[0, 0].plot(t, y1, 'r-', label="sin(t)")  # 'r-' stands for a red solid line
axs[0, 0].set_title("Basic Sin Wave 🎵")  # Set the title for this subplot
axs[0, 0].legend()  # Display the legend to identify the plot

# 📈 Step 6: Plot on the second subplot (top-right)
axs[0, 1].plot(t, y2, 'g--', label="sin(2t)")  # 'g--' indicates a green dashed line
axs[0, 1].set_title("Double Frequency Sin 🌊")  # Title indicating higher frequency
axs[0, 1].legend()  # Display the legend

# 📈 Step 7: Plot on the third subplot (bottom-left)
axs[1, 0].plot(t, y3, 'b-.', label="sin(3t)")  # 'b-.' stands for a blue dash-dot line
axs[1, 0].set_title("Triple Frequency Sin ⚡")  # Title for triple frequency
axs[1, 0].legend()  # Show legend for better clarity

# 📈 Step 8: Plot all waves together in the fourth subplot (bottom-right)
# This subplot combines all the sine waves for comparison
axs[1, 1].plot(t, y1, 'r-', label="sin(t)")     # Standard sine wave
axs[1, 1].plot(t, y2, 'g--', label="sin(2t)")   # Double frequency sine wave
axs[1, 1].plot(t, y3, 'b-.', label="sin(3t)")   # Triple frequency sine wave
axs[1, 1].set_title("All Waves Together 🎶")    # Title for combined plot
axs[1, 1].legend()  # Legend shows all plot labels for clarity

# 🚀 Step 9: Adjust layout for better spacing between plots
plt.tight_layout()  # Automatically adjusts subplot parameters to fit the figure area

# 🎯 Step 10: Display the complete set of plots
plt.show()  # Render the plots and display them
```

# ADJUST FIGURE SIZE, SPACING, COLOR

- plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.4)
    - left: left side of the subplots of the figure
    - right: right side of the subplots of the figure
    - bottom: bottom of the subplots of the figure
    - top: top of the subplots of the figure
    - wspace: amount of width reserved for space between subplots, expressed as a fraction of the average axis width
    - hspace: amount of height reserved for space between subplots, expressed as a fraction of the average axis height

```python
import numpy as np
import matplotlib.pyplot as plt

# 🎯 Generate x values (100 points from 0 to 10)
x = np.linspace(0, 10, 100)  

# 📈 Create three sine wave functions
y1 = np.sin(x)       # 🔴 Standard sine wave
y2 = np.sin(2 * x)   # 🔵 Double frequency sine wave
y3 = np.sin(3 * x)   # 🟢 Triple frequency sine wave

# 🎨 Create a figure with 3 vertically stacked subplots
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 10))  # Set figure size to 10x10

# 🔴 First plot - Standard sine wave
axes[0].plot(x, y1, 'r')   # 'r' → red line
axes[0].set_xlabel('x1')   # Label for X-axis
axes[0].set_ylabel('y1')   # Label for Y-axis
axes[0].set_title('📌 Title 1 - Standard Sin Wave')

# 🔵 Second plot - Double frequency sine wave
axes[1].plot(x, y2, 'b')   # 'b' → blue line
axes[1].set_xlabel('x2')
axes[1].set_ylabel('y2')
axes[1].set_title('📌 Title 2 - Double Frequency Sin Wave')

# 🟢 Third plot - Triple frequency sine wave
axes[2].plot(x, y3, 'g')   # 'g' → green line
axes[2].set_xlabel('x3')
axes[2].set_ylabel('y3')
axes[2].set_title('📌 Title 3 - Triple Frequency Sin Wave')

# ✨ Adjust spacing between subplots for better readability
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.4)

# 🚀 Display the final figure
plt.show()
```

# 3-D PLOTTING

```python
# 📌 Importing necessary libraries
import numpy as np  # 🔢 Numerical operations
import matplotlib.pyplot as plt  # 📊 Plotting library
from mpl_toolkits.mplot3d import Axes3D  # 📈 3D plotting module

# 🔷 Step 1: Creating a Simple 3D Scatter Plot 📍
fig = plt.figure(figsize=(6, 6))  # 📏 Setting figure size
ax = fig.add_subplot(111, projection='3d')  # 🏗️ Adding 3D subplot

# 🎯 Generate random 3D points
x = np.random.rand(10) * 10  # 📊 X-coordinates
y = np.random.rand(10) * 10  # 📊 Y-coordinates
z = np.random.rand(10) * 10  # 📊 Z-coordinates

ax.scatter(x, y, z, color='red', marker='o', s=50)  # 🔴 Plot red points
ax.set_title("📍 3D Scatter Plot Example")  # 🏷️ Title

# 🏷️ Setting axis labels
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()  # 🚀 Display the plot


# 🔷 Step 2: Creating a 3D Wireframe Plot 📡
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')

# 🎯 Generate mesh grid for surface
X = np.linspace(-5, 5, 50)  # 📏 X values
Y = np.linspace(-5, 5, 50)  # 📏 Y values
X, Y = np.meshgrid(X, Y)  # 🏗️ Creating a 2D grid
Z = np.sin(np.sqrt(X**2 + Y**2))  # 📈 Compute Z values

ax.plot_wireframe(X, Y, Z, color='blue')  # 🔵 Create wireframe plot
ax.set_title("📡 3D Wireframe Plot Example")  # 🏷️ Title

# 🏷️ Labeling axes
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()  # 🚀 Display the plot


# 🔷 Step 3: Creating a 3D Surface Plot 🌋
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')

# 🎯 Generate mesh grid for 3D surface
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 🎨 Creating a surface plot with color mapping
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')

ax.set_title("🌋 3D Surface Plot with Color Mapping")  # 🏷️ Title

# 🏷️ Labeling axes
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.colorbar(surf)  # 🎨 Adding a color bar for reference
plt.show()  # 🚀 Display the plot
```

```python
# 📌 Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 🏗️ 3D plotting module

# 🎯 Step 1: Generate a meshgrid for 3D plotting
X = np.linspace(-5, 5, 100)  # 🔵 X-axis values (ranging from -5 to 5)
Y = np.linspace(-5, 5, 100)  # 🔵 Y-axis values (ranging from -5 to 5)
X, Y = np.meshgrid(X, Y)  # 📏 Creating a grid
Z = np.sin(np.sqrt(X**2 + Y**2)) * np.cos(X)  # 🌊 Applying a custom function

# 🎨 Step 2: Create the 3D figure
fig = plt.figure(figsize=(10, 7))  # 📏 Set figure size
ax = fig.add_subplot(111, projection='3d')  # 🎛️ Add a 3D subplot

# 🔷 Step 3: Plot the wireframe with more details
ax.plot_wireframe(X, Y, Z, color='purple', linewidth=1.2, linestyle="dashed")  # 🔗 Dashed wireframe

# ✨ Adding a scatter plot for extra details
ax.scatter(0, 0, 0, color='red', s=100, label="Origin Point")  # 📍 Highlighting origin

# 🏷️ Step 4: Customize axis labels and title
ax.set_xlabel("📏 X-axis (Range: -5 to 5)", fontsize=12, color="blue")
ax.set_ylabel("📏 Y-axis (Range: -5 to 5)", fontsize=12, color="green")
ax.set_zlabel("📏 Z-axis (Function Output)", fontsize=12, color="red")
ax.set_title("🌍 Advanced 3D Wireframe with Customization", fontsize=14, fontweight="bold")

# 🎨 Step 5: Adjust viewing angle for better visualization
ax.view_init(elev=30, azim=45)  # 🔄 Rotate the view (Elevation: 30°, Azimuth: 45°)

# 📢 Show the legend
ax.legend()

# 🚀 Step 6: Display the final plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importing 3D plotting toolkit 🛠️

# Generate data for the plots 📊
X = np.linspace(-5, 5, 50)  # X-axis values ranging from -5 to 5
Y = np.linspace(-5, 5, 50)  # Y-axis values ranging from -5 to 5
X, Y = np.meshgrid(X, Y)  # Create a grid of X and Y coordinates
Z = np.sin(np.sqrt(X**2 + Y**2))  # Define the Z values using a wave function 🌊

# Create a figure for multiple subplots 🎨
fig = plt.figure(figsize=(14, 10))  # Set figure size

# 🌄 1st Plot: 3D Surface Plot - Smooth surface visualization
ax1 = fig.add_subplot(221, projection='3d')  # Create a 3D subplot
ax1.plot_surface(X, Y, Z, cmap='viridis')  # Use 'viridis' colormap for aesthetics 🎨
ax1.set_title("3D Surface Plot 🌄")  # Set title

# 🌀 2nd Plot: 3D Contour Plot - Contour lines in 3D
ax2 = fig.add_subplot(222, projection='3d')  # Create another 3D subplot
ax2.contour3D(X, Y, Z, 50, cmap='coolwarm')  # Generate contour lines with 50 levels
ax2.set_title("3D Contour Plot 🌀")  # Set title

# 🔴🟢🔵 3rd Plot: 3D Scatter Plot - Random points in space
ax3 = fig.add_subplot(223, projection='3d')  # Create another 3D subplot
x_scatter = np.random.rand(100) * 10 - 5  # Generate random x-coordinates
y_scatter = np.random.rand(100) * 10 - 5  # Generate random y-coordinates
z_scatter = np.random.rand(100) * 10 - 5  # Generate random z-coordinates
ax3.scatter(x_scatter, y_scatter, z_scatter, c=z_scatter, cmap='rainbow')  # Use 'rainbow' colormap 🌈
ax3.set_title("3D Scatter Plot 🔴🟢🔵")  # Set title

# 📊 4th Plot: 3D Bar Chart - Vertical bars in 3D
ax4 = fig.add_subplot(224, projection='3d')  # Create another 3D subplot
_x = np.arange(4)  # X-axis positions for bars
_y = np.arange(3)  # Y-axis positions for bars
_xx, _yy = np.meshgrid(_x, _y)  # Create a grid for bars
x, y = _xx.ravel(), _yy.ravel()  # Flatten the grid arrays
top = np.random.rand(len(x))  # Generate random heights for bars
bottom = np.zeros_like(top)  # Set base height to zero
width = depth = 0.5  # Set bar width and depth
ax4.bar3d(x, y, bottom, width, depth, top, shade=True)  # Plot 3D bars
ax4.set_title("3D Bar Plot 📊")  # Set title

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots 🎉
plt.show()
```

# PIE PLOTS

```python
import matplotlib.pyplot as plt

# 🎯 Data for the pie chart
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 25, 35, 10]  # 🍏🍌🍒🌰 Proportion of each category
colors = ['#FF9999', '#FFD700', '#8A2BE2', '#20B2AA']  # 🎨 Custom colors

# 🥧 Creating the pie chart
plt.figure(figsize=(6, 6))  # 📏 Adjust figure size
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# 📢 Adding a title
plt.title("🍏 Fruit Distribution Pie Chart", fontsize=14, fontweight="bold")

# 📌 Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# 🎯 Data for the pie chart
labels = ['USA', 'India', 'China', 'Germany', 'Brazil']
sizes = [40, 20, 15, 10, 15]  # 🌎 Distribution by country
colors = ['red', 'blue', 'green', 'purple', 'orange']  # 🎨 Different colors
explode = (0.1, 0, 0, 0, 0.2)  # 💥 Highlighting USA & Brazil

# 🥧 Creating the pie chart
plt.figure(figsize=(7, 7))  # 📏 Set figure size
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', 
        explode=explode, shadow=True, startangle=90)

# 📢 Adding a title
plt.title("🌎 Global Population Share", fontsize=14, fontweight="bold")

# 📌 Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# 🎯 Data for the donut chart
labels = ['Tech', 'Healthcare', 'Finance', 'Retail']
sizes = [40, 30, 20, 10]  # 📊 Industry revenue distribution
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff6666']  # 🎨 Custom pastel colors

# 🥧 Creating the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# 🏆 Creating the "donut hole"
center_circle = plt.Circle((0, 0), 0.5, fc='white')  # ⚪ Add a circle in the middle
plt.gca().add_artist(center_circle)

# 📢 Adding a title
plt.title("💰 Industry Revenue Share (Donut Chart)", fontsize=14, fontweight="bold")

# 📌 Display the plot
plt.show()
```
