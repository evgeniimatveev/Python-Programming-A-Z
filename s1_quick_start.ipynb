# S1 - Quick Start

Small warm-up notebook for **Python Programming A-Z**.

In this notebook you will:
- define a few variables,
- store lesson names in a list,
- print them with a simple `for` loop.


```python
# Simple warm-up: variables and a loop
course = 'Python Programming A-Z'
section = 'S1 - Python Basics'
lessons = ['Variables', 'Strings', 'Lists']

print(course)
print(section)
print('Lessons in this section:')

for i, name in enumerate(lessons, start=1):
    print(f'{i}. {name}')
```
