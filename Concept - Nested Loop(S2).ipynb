# NESTED LOOPS
- Nested loops are loops that exist inside the body of another loop

```python
# ===========================================================
# PYTHON PROGRAMMING FUNDAMENTALS: NESTED LOOPS üöÄ
# ===========================================================

# Introduction:
# --------------
# Nested loops are loops inside the body of another loop.
# These are useful when you need to iterate through multiple dimensions,
# such as rows and columns in a matrix or elements in combinations.

# Let's start exploring nested loops with examples! üéØ

# ===========================================================
# Example 1: Printing all pairs of (i, j)
# ===========================================================
print("Example 1: Generating all pairs of numbers (i, j) where i ranges from 1 to 3 and j ranges from 1 to 3:")

for i in range(1, 4):  # Outer loop: Controls the first number
    for j in range(1, 4):  # Inner loop: Controls the second number
        print(f"({i}, {j})")  # Print each pair
# Output:
# (1, 1)
# (1, 2)
# (1, 3)
# (2, 1)
# (2, 2)
# (2, 3)
# (3, 1)
# (3, 2)
# (3, 3)

# Explanation:
# 1Ô∏è‚É£ The outer loop iterates over values 1 to 3.
# 2Ô∏è‚É£ For each iteration of the outer loop, the inner loop runs completely (from 1 to 3).
# 3Ô∏è‚É£ This results in printing all possible pairs of (i, j).  


# ===========================================================
# Example 2: Nested loops with conditions
# ===========================================================
print("\nExample 2: Generating a right-angle triangle pattern using nested loops:")

n = 5  # Height of the triangle
for i in range(1, n + 1):  # Outer loop: Controls the number of rows
    for j in range(i):  # Inner loop: Controls the number of stars in each row
        print("*", end="")  # Print stars in the same line
    print()  # Move to the next line after completing a row
# Output:
# *
# **
# ***
# ****
# *****

# Explanation:
# 1Ô∏è‚É£ The outer loop runs `n` times, where `n` is the height of the triangle.
# 2Ô∏è‚É£ For each iteration of the outer loop, the inner loop runs `i` times to print the required number of stars.
# 3Ô∏è‚É£ After each row of stars is printed, the `print()` moves the cursor to the next line.


# ===========================================================
# Example 3: Nested loops to combine two lists
# ===========================================================
print("\nExample 3: Combining elements of two lists using nested loops:")

colors = ['blue', 'red', 'green']
objects = ['car', 'ball', 'chair']

for color in colors:  # Outer loop: Iterates over the first list (colors)
    for obj in objects:  # Inner loop: Iterates over the second list (objects)
        print(f"{color} {obj}")
# Output:
# blue car
# blue ball
# blue chair
# red car
# red ball
# red chair
# green car
# green ball
# green chair

# Explanation:
# 1Ô∏è‚É£ For each `color` in the colors list, the inner loop runs completely through the objects list.
# 2Ô∏è‚É£ This results in all possible combinations of colors and objects.


# ===========================================================
# BONUS: Multiplication Table
# ===========================================================
print("\nBONUS: Generating a multiplication table using nested loops:")

n = 5  # Up to 5x5 multiplication table
for i in range(1, n + 1):  # Outer loop: Represents rows
    for j in range(1, n + 1):  # Inner loop: Represents columns
        print(f"{i * j:2}", end=" ")  # Print the product, formatted for alignment
    print()  # Move to the next row
# Output:
#  1  2  3  4  5
#  2  4  6  8 10
#  3  6  9 12 15
#  4  8 12 16 20
#  5 10 15 20 25

# Explanation:
# 1Ô∏è‚É£ The outer loop iterates through the rows (numbers 1 to n).
# 2Ô∏è‚É£ The inner loop iterates through the columns, calculating the product `i * j` for each pair.
# 3Ô∏è‚É£ The `end=" "` keeps the values in the same row until the inner loop finishes.


# --- SUMMARY ---
# ---------------
# ‚úÖ Nested loops allow us to handle multi-dimensional data effectively.
# ‚úÖ They are helpful for generating combinations, patterns, and tables.
# ‚úÖ Be cautious with nested loops for large datasets‚Äîthey can be computationally expensive.
# ‚úÖ Practice using nested loops with conditions and combinations for mastery. üí™
```
